```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical

# --- Helper function to load the raw sequence files ---

def load_sequences(filepath, set_name):
    """
    Loads the 9 sequence files (X, Y, Z for total, body, and gyro)
    and stacks them into a single 3D array.
    """
    signals = []
    
    # The 9 files to load
    signal_files = [
        'total_acc_x_', 'total_acc_y_', 'total_acc_z_',
        'body_acc_x_', 'body_acc_y_', 'body_acc_z_',
        'body_gyro_x_', 'body_gyro_y_', 'body_gyro_z_'
    ]
    
    print(f"Loading raw sequences for {set_name}...")
    for sig_file in signal_files:
        filename = f"{filepath}{set_name}/Inertial Signals/{sig_file}{set_name}.txt"
        signal_data = pd.read_csv(filename, delim_whitespace=True, header=None).values
        signals.append(signal_data)
        
    # Stack the 9 signals (features) on top of each other
    # The shape becomes (num_samples, 128_timesteps, 9_features)
    return np.stack(signals, axis=-1)

# --- 1. Load the Raw Sequence Data ---
# Note: This is a different X_train from your previous models!
data_path = 'UCI HAR Dataset/' 

X_train_lstm = load_sequences(data_path, 'train')
X_test_lstm = load_sequences(data_path, 'test')

# X_train_lstm.shape will be (7352, 128, 9)
# 7352 samples, 128 timesteps, 9 features (sensors)
print(f"X_train shape: {X_train_lstm.shape}") 
print(f"X_test shape: {X_test_lstm.shape}")


# --- 2. Load and Prepare the Labels (y) ---
# We load the same y_train/y_test files, but must make one key change

y_train_df = pd.read_csv(f"{data_path}train/y_train.txt", header=None)
y_test_df = pd.read_csv(f"{data_path}test/y_test.txt", header=None)

# Keras needs labels to be 0-indexed (0 to 5)
# Your labels are 1-indexed (1 to 6). We MUST subtract 1.
y_train_lstm = y_train_df[0] - 1
y_test_lstm = y_test_df[0] - 1

# We will NOT one-hot encode. We'll use 'sparse_categorical_crossentropy'
# which is simpler and handles integer labels.


# --- 3. Build the Keras LSTM Model ---
# We can now define the neural network architecture

n_timesteps = X_train_lstm.shape[1]  # 128
n_features = X_train_lstm.shape[2]   # 9
n_classes = 6                        # 6 activities

model = Sequential()

# Input LSTM layer
# 100 is the number of LSTM units (a good starting point)
model.add(LSTM(100, input_shape=(n_timesteps, n_features)))

# Dropout layer to prevent overfitting
model.add(Dropout(0.5))

# A standard fully-connected (Dense) layer
model.add(Dense(100, activation='relu'))

# Output layer
# 6 units (one for each class)
# 'softmax' ensures the output is a probability distribution
model.add(Dense(n_classes, activation='softmax'))

model.summary()


# --- 4. Compile the Model ---

# Compiling configures the model for training
model.compile(
    loss='sparse_categorical_crossentropy', # Use this for integer labels
    optimizer='adam',                       # A great default optimizer
    metrics=['accuracy']
)


# --- 5. Train the Model ---
# This is where the GPU does all the work!
print("\nTraining LSTM model...")

# Set training parameters
epochs = 20
batch_size = 64 # Number of samples to process at one time

# Start training
# We use X_test_lstm/y_test_lstm as our validation data
history = model.fit(
    X_train_lstm, 
    y_train_lstm,
    epochs=epochs,
    batch_size=batch_size,
    validation_data=(X_test_lstm, y_test_lstm),
    verbose=1 # Shows the progress bar
)

print("Training complete.")


# --- 6. Evaluate and Get Confusion Matrix ---

# 1. Plot training history
pd.DataFrame(history.history).plot(figsize=(10, 6))
plt.title('Model Training History')
plt.ylabel('Accuracy / Loss')
plt.xlabel('Epoch')
plt.show()

# 2. Get final accuracy on the test set
test_loss, test_accuracy = model.evaluate(X_test_lstm, y_test_lstm, batch_size=batch_size)
print(f"\nTest Accuracy: {test_accuracy*100:.2f}%")

# 3. Get the confusion matrix
y_pred_probs = model.predict(X_test_lstm)
y_pred_classes = np.argmax(y_pred_probs, axis=1) # Get the class with highest prob

# Use scikit-learn's confusion_matrix
cm = confusion_matrix(y_test_lstm, y_pred_classes)

# Plot using your 'plot_confusion_matrix' function
# (Make sure 'labels' is defined as your list of activity names)
labels = ['LAYING', 'SITTING', 'STANDING', 'WALKING', 'WALKING_DOWNSTAIRS', 'WALKING_UPSTAIRS']

plt.figure(figsize=(7,7))
# Assuming you have your plot_confusion_matrix function from before
# and that it's imported or defined in another cell
try:
    plot_confusion_matrix(cm, classes=labels, normalize=True, title='LSTM Confusion Matrix')
    plt.show()
except NameError:
    print("\n'plot_confusion_matrix' function not found. Printing text matrix:")
    print(cm)
```
